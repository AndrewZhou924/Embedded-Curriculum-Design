```python
file = "./data/output.txt"
f    = open(file, 'r')
```

```python
f.readlines()
```

```python
all_string = ""
f          = open(file, 'r')
for line in f.readlines(): 
    if '2020' in line:
        continue
    all_string += line.strip('\n')
```

```python
len(all_string)
```

```python
each_lines = all_string.split('AE 00')
```

```python
each_lines = each_lines[:-1]
```

```python
len(each_lines)
```

```python
each_lines[0].split(' ')[-1]
```

```python
data_np = np.zeros((480, 320, 3))
```

```python
data_np.shape
```

```python
cnt1 = 0
cnt2 = 0
for i,line in enumerate(each_lines):
    split_line = line.split(' ')[:-1]
    for j in range(640)[::2]:
        pixel = split_line[j]+split_line[j+1]

        if pixel=="FFFF":
            data_np[i, int(j/2), :] = np.array([255, 255, 255])
            cnt1 += 1
        else:
            data_np[i, int(j/2), :] = np.array([0,   0,   0])
            cnt2 += 1
```
